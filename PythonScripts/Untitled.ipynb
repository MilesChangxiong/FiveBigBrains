```python
import requests
import pandas as pd
import matplotlib.pyplot as plt

# Your Firebase database URL and access key (keep it secure)
firebase_url = "https://fivebigbrains-usccs526fall2023-default-rtdb.firebaseio.com/"
access_key = ""
```

```python
REMOVE_DEV_DATA = False
```

```python
# Get data from Firebase
response = requests.get(f"{firebase_url}/mapData.json?auth={access_key}")
data = response.json()

# Initialize lists to collect statistics
map_names = []
movement_distances = []
attack_counts = []
jump_counts = []
taunt_counts = []

# Loop through data and collect statistics
for key, value in data.items():
    if REMOVE_DEV_DATA and value.get('Environment') == 'UnityEditor':
        continue
    if not value.get('mapName'):
        continue
    if value.get('mapName') == 'Tutorial':
        continue
    statistics = value.get('statistics', {})
    map_names.append(value.get('mapName', 'Unknown'))
    movement_distances.append(statistics.get('MovementDistance', 0))
    attack_counts.append(statistics.get('AttackCount', 0))
    jump_counts.append(statistics.get('JumpCount', 0))
    taunt_counts.append(statistics.get('TauntCount', 0))

# Create DataFrame from collected data
df = pd.DataFrame({
    'Map': map_names,
    'MovementDistance': movement_distances,
    'AttackCount': attack_counts,
    'JumpCount': jump_counts,
    'TauntCount': taunt_counts
})

# Calculate average for each map
average_df = df.groupby('Map', as_index=False).mean()

# Set the figure size
plt.figure(figsize=(10, 6))

# Create primary axis for movement distances
ax1 = plt.subplot(1, 1, 1)
bar_width = 0.2
# Numeric index for the position of the groups
numeric_index = range(len(average_df))

# Plot Movement Distance on primary y-axis
bars1 = plt.bar(numeric_index, average_df['MovementDistance'], bar_width, label='MovementDistance', color='b')

# Create secondary axis for counts
ax2 = ax1.twinx()

# Plot Attack Count on secondary y-axis
bars2 = ax2.bar([x + bar_width for x in numeric_index], average_df['AttackCount'], bar_width, label='AttackCount', color='r')

# Plot Jump Count on secondary y-axis
bars3 = ax2.bar([x + 2*bar_width for x in numeric_index], average_df['JumpCount'], bar_width, label='JumpCount', color='g')

# Plot Taunt Count on secondary y-axis
bars4 = ax2.bar([x + 3*bar_width for x in numeric_index], average_df['TauntCount'], bar_width, label='TauntCount', color='y')

# Set legend
# Set legend with adjusted position and added transparency
ax1.legend((bars1[0], bars2[0], bars3[0], bars4[0]), ('MovementDistance', 'AttackCount', 'JumpCount', 'TauntCount'), 
           loc='upper right', bbox_to_anchor=(1, 1), fancybox=True, shadow=True, facecolor='white', framealpha=0.7)


# Set chart title and labels
ax1.set_title('Game Statistics by Map')
ax1.set_xlabel('Map')
ax1.set_ylabel('MovementDistance', color='b')
ax2.set_ylabel('Counts', color='r')

# Set label colors for each y-axis
ax1.tick_params('y', colors='b')
ax2.tick_params('y', colors='r')

# Set the position and labels of the x-ticks
plt.xticks([x + 1.5*bar_width for x in numeric_index], average_df['Map'], rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Get weapon event data from Firebase
response = requests.get(f"{firebase_url}/weaponEvent.json?auth={access_key}")
data = response.json()

weapon_bullet_mapping = {
    'FireBullet': 'FireBallShooter',
    'MagnifyBullet': 'MagnifyGun',
    'PistolBullet': 'Pistol'
}
weapon_stats = []

for event in data.values():
    if REMOVE_DEV_DATA and event.get('Environment') == 'UnityEditor':
        continue
        
    weapon_name = event.get('weaponName')
    event_type = event.get('eventType')
    
    if event_type == 'HitIce':
        continue

    weapon_name = weapon_bullet_mapping.get(weapon_name, weapon_name)
    
    weapon_stats.append([weapon_name, event_type])

df = pd.DataFrame(weapon_stats, columns=['WeaponName', 'EventType'])

# Determine hit events
hit_events = df[df['EventType'].str.startswith('Hit')]

# Count occurrences for each weapon and event type
weapon_event_counts = hit_events.groupby(['WeaponName', 'EventType']).size().unstack(fill_value=0)

# Add Attack count and compute Missed count
total_attacks = df[df['EventType'] == 'Attack'].groupby('WeaponName').size()
weapon_event_counts['Missed'] = total_attacks - weapon_event_counts.sum(axis=1)
weapon_event_counts = weapon_event_counts.div(total_attacks, axis=0) * 100  # Convert to percentages

# Plot
fig, ax = plt.subplots(figsize=(12, 6))
weapon_event_counts.plot(kind='bar', stacked=True, ax=ax)
plt.title('Weapon Hit/Miss Proportions')
plt.ylabel('Percentage (%)')
plt.xlabel('Weapon Name')
plt.ylim(0, 100)  # Ensure y-axis goes from 0% to 100%
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
# Get weapon event data from Firebase
response = requests.get(f"{firebase_url}/weaponEvent.json?auth={access_key}")
data = response.json()

weapon_bullet_mapping = {
    'FireBullet': 'FireBallShooter',
    'MagnifyBullet': 'MagnifyGun',
    'PistolBullet': 'Pistol'
}
weapon_stats = []

for event in data.values():
    weapon_name = event.get('weaponName')
    event_type = event.get('eventType')
    is_taunted = event.get('isOpponentTaunted')
    
    if event_type == 'HitIce':
        continue
    
    weapon_name = weapon_bullet_mapping.get(weapon_name, weapon_name)
    
    if weapon_name == 'MagnifyGun':
        continue
    
    weapon_stats.append([weapon_name, event_type, is_taunted])

df = pd.DataFrame(weapon_stats, columns=['WeaponName', 'EventType', 'IsOpponentTaunted'])

# Create 'Hit' column
df['Hit'] = df['EventType'].apply(lambda x: 1 if x.startswith('Hit') else 0)

# Compute hit rates
hit_rates = df.groupby(['WeaponName', 'IsOpponentTaunted'])['Hit'].mean().unstack()

# Convert hit rates to percentages
hit_rates *= 100

# Plot
fig, ax = plt.subplots(figsize=(12, 6))
hit_rates.plot(kind='bar', ax=ax)
plt.title('Hit Rate by Weapon and Taunt Status')
plt.ylabel('Hit Rate (%)')
plt.xlabel('Weapon Name')
plt.ylim(0, 100)  # Ensure y-axis goes from 0% to 100%
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(title='Is Opponent Taunted')
plt.show()
```

```python
# Retrieve data from Firebase's /mapData
response = requests.get(f"{firebase_url}/mapData.json?auth={access_key}")
data = response.json()

# Extract data
stats = []
for entry in data.values():
    if REMOVE_DEV_DATA and entry.get('Environment') == 'UnityEditor':
        continue
    if entry.get('mapName') != 'FiregunAndIce':
        continue
        
    statistics = entry["statistics"]
    
    p1_pickups = statistics["Player1PowerupPickupCount"]
    p2_pickups = statistics["Player2PowerupPickupCount"]
    
    p1_wins = statistics["Player1Wins"]
    p2_wins = statistics["Player2Wins"]
    
    stats.append([p1_pickups, p1_wins])
    stats.append([p2_pickups, p2_wins])

# Create a DataFrame from the stats
df = pd.DataFrame(stats, columns=["PowerupPickups", "Wins"])

# Calculate win rates
win_rates = df.groupby("PowerupPickups")["Wins"].mean() * 100

# Plotting the chart
fig, ax = plt.subplots(figsize=(10, 6))
win_rates.plot(kind='bar', ax=ax, color='blue')
plt.title('Relationship Between Win Rate and Powerup Pickup Count in FiregunAndIce')
plt.ylabel('Win Rate (%)')
plt.xlabel('Powerup Pickup Count')
plt.ylim(0, 100)  # Set y-axis limits to be from 0 to 100%
plt.xticks(rotation=0)  # Keep the x-axis labels horizontal
plt.tight_layout()  # Adjust the padding between and around subplots
plt.show()
```
